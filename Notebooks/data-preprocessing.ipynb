## Converting columns to sensible names

```python
data.rename(columns={
    'Unnamed: 0': 'Index',
    'SEQN': 'Respondent ID',
    'RIAGENDR': 'Gender',
    'RIDAGEYR': 'Age',
    'DMDHHSIZ': 'Household Size',
    'INDFMPIR': 'Income Poverty Ratio',
    'BMXBMI': 'BMI',
    'DSD010': 'Diet Question',
    'DSD010AN': 'Diet Question Annotation',
    'SMD415': 'Smoking Status',
    'PAD590': 'Physical Activity Frequency',
    'PAD600': 'Physical Activity Duration',
    'HUQ010': 'General Health Status',
    'restaurant': 'Restaurant Visits',
    'protein': 'Protein Intake',
    'healthy': 'Healthy Food Intake',
    'unhealthy': 'Unhealthy Food Intake',
    'beverage': 'Beverage Consumption',
    'milk': 'Milk Consumption',
    'MCQ010': 'Medical Condition 1',
    'MCQ053': 'Medical Condition 2',
    'MCQ092': 'Medical Condition 3',
    'MCQ140': 'Medical Condition 4',
    'active': 'Physical Activity Status'
}, inplace=True)


data.head()
```

## Chossing a subset of columns to work with

## Universal Division or Train and Test Sets
- 60 - TRAIN
- 40 - TEST

- Trying to create a situation of less amount of data present

```python
from sklearn.model_selection import train_test_split

# Define target column
target_column = 'Physical Activity Status'

# Split the data
train_data, test_data = train_test_split(data, test_size=0.4, random_state=42, stratify=data[target_column])

# Save to CSV
train_data.to_csv('../data/universal_train_data.csv', index=False)
test_data.to_csv('../data/universal_test_data.csv', index=False)
```
